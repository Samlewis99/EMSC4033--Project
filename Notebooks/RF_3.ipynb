# Visualize our model and find important features
We now have a random forest model, but we cant see anything of it! Lets import some othe sci kit learn packages to both visualise our model, and see what features are most signigicant in the classification.

Random forest uses gini importance or mean decrease in impurity (MDI) to calculate the importance of each feature. This is how much the model fit or accuracy decreases when you drop a variable. The larger the decrease, the more significant the variable is.

We will call a function to bring our data and model into this notebook.

```python
from Functions_r import f_rf as ft
clf,iris = ft.get_data1()
```

```python
#Find variable importance
import pandas as pd
feature_imp = pd.Series(clf.feature_importances_,index=iris.feature_names).sort_values(ascending=False)
feature_imp
```

## Visualise the important features
We often need to be able to graphically describe our results for reports or for consumers, stakeholders to see. We can use matplotlib to graph our vairiable importance to save for later.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Creating a bar plot
sns.barplot(x=feature_imp, y=feature_imp.index)
# Add labels to your graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```

In the next module we will use our most important features to train our random forest model to see if we can improve classification rate, and decrease processing time

## Visualise the decision tree
A interesting package in the scikit-learn library allows us to actually visualise a single tree in our random forest. Although we can't get the full picture of the forest it is useful in understanding how the decision tree works to classify a variable.


```python
# Import tools needed for visualization
from sklearn.tree import plot_tree
plt.figure(figsize=(80,40))
plot_tree(clf.estimators_[3],feature_names=iris.feature_names, class_names = iris.target_names, filled=True);
```
